```bash
git clone https://github.com/brielin/Popcorn.git
cd Popcorn
pip install .
```

```bash
popcorn -h   # check popcorn was installed correctly
```

# Run Popcorn

```bash
popcorn fit -h    # shows fit usage
```

```bash
PARENT_DIR=/mnt/project/Aim4_CrossAncestryGeneticCorrelation/jahagirdarob
SCORES_PREF=${PARENT_DIR}/scores_files
SUMSTATS_PREF=${PARENT_DIR}/sumstats_files
```

## Overall Breast Cancer Summary Statistics

### African and East Asian populations

```bash
popcorn fit -v 1 --sfile1 $SUMSTATS_PREF/AFR_meta.sumstats.txt --sfile2 $SUMSTATS_PREF/EAS_meta.sumstats.txt --cfile $SCORES_PREF/AFR_EAS_scores.txt AFR_EAS_meta_corr.txt
```
