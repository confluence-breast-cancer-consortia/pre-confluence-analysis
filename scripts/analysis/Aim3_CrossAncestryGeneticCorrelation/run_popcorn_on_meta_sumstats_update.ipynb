```bash
SCORES_PREF=/mnt/project/Aim4_CrossAncestryGeneticCorrelation/jahagirdarob/scores_files
SUMSTATS_PREF=/mnt/project/Aim4_CrossAncestryGeneticCorrelation/jahagirdarob/sumstats_files
```

```bash
dx ls ../scores_files
```

```bash
# SUBSET="HM3_MEGA"
SUBSET="HM3"
POPS="AFR AMR EAS EUR"

read -r -a POPSARR <<< "$POPS"
NUMPOPS=${#POPSARR[@]}

for ((i = 0; i < NUMPOPS; i++))
do
    POP1=${POPSARR[i]}
    MAF1="01"
    if [ "$POP1" = "AMR" ]
    then
        MAF1="05"
    fi
    
    for ((j = i + 1; j < NUMPOPS; j++))
    do
        POP2=${POPSARR[j]}
        MAF2="01"
        if [ "$POP2" = "AMR" ]
        then
            MAF2="05"
        fi

        if [ "$POP1" != "AMR" ] && [ "$POP2" != "AMR" ]
        then
            break
        fi

        popcorn fit -v 1 \
           --sfile1 $SUMSTATS_PREF/${POP1}_Overall_${SUBSET}_${MAF1}.sumstats.txt \
           --sfile2 $SUMSTATS_PREF/${POP2}_Overall_${SUBSET}_${MAF2}.sumstats.txt \
           --cfile $SCORES_PREF/${POP1}_${POP2}_cleaned_scores.txt \
           ${POP1}_${POP2}_Overall_${SUBSET}.txt
        
        dx upload ${POP1}_${POP2}_Overall_${SUBSET}.txt*
    done
done
```
