```python
corr_dir = "/mnt/project/Aim4_CrossAncestryGeneticCorrelation/jahagirdarob/corr_files"
```

```python
def get_pops_corr_df(pops_corr_dir, idx_name, col_pref="", gen=False):
    corr_label = "pge" if gen else "pgi"
    pops_corr_df = pd.read_csv(pops_corr_dir, sep='\t', index_col=0).loc[[corr_label]]
    pops_corr_df.index = [idx_name]
    pops_corr_df.columns = [f"{col_pref}{col.split()[0]}" for col in pops_corr_df.columns]
    return pops_corr_df
```

```python
pop_pair_corr_df_dict = {}

overall_pops = ["AFR", "AMR", "EAS", "EUR"]

subset = "HM3"
#subset = "HM3_MEGA"

maf_suffix = ""
#maf_suffix = "_05"

for i, pop1 in enumerate(overall_pops):
    for pop2 in overall_pops[i + 1:]:
        pop_pair_str = f"{pop1}_{pop2}"
        pop_pair_corr_dir = f"{corr_dir}/{pop_pair_str}_Overall_{subset}{maf_suffix}.txt"
        pop_pair_corr_df = get_pops_corr_df(pop_pair_corr_dir, pop_pair_str)
        
        if pop_pair_str not in pop_pair_corr_df_dict:
            pop_pair_corr_df_dict[pop_pair_str] = []
        pop_pair_corr_df_dict[pop_pair_str].append(pop_pair_corr_df)
        
pop_pair_corr_dfs = []
for pop_pair_str in pop_pair_corr_df_dict:
    pop_pair_corr_dfs.append(pd.concat(pop_pair_corr_df_dict[pop_pair_str], axis=1))
corr_df = pd.concat(pop_pair_corr_dfs)
corr_df.to_csv(f"Overall_{subset}{maf_suffix}_results.txt", sep='\t')
corr_df
```
